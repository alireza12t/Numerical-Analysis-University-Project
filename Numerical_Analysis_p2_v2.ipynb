```
from math import pi, cos
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
```

```
def chebyshevPoints(k):
  answ = []
  for i in range(1 , k+1):
    x_k = cos((((2*i) - 1)/(2*k))* pi )
    answ.append(x_k)
  res = np.asarray(answ)
  return res
```

```
chebyshevPoints(10)
```

```
x5 = chebyshevPoints(6)
x10 = chebyshevPoints(11)
x15 = chebyshevPoints(16)
x20 = chebyshevPoints(21)
y5 = 1 / (1+ 25*(x5**2))
y10 = 1 / (1+ 25*(x10**2))
y15 = 1 / (1+ 25*(x15**2))
y20 = 1 / (1+ 25*(x20**2))




p5 = np.polyfit(x5 , y5 , len(x5) - 1)
p10 = np.polyfit(x10 , y10 , len(x10) - 1)
p15 = np.polyfit(x15 , y15 , len(x15) - 1)
p20 = np.polyfit(x20 , y20 , len(x20) - 1)

x_min = min(list(x10) + list(x15) + list(x15) + list(x20))
x_max = max(list(x10) + list(x15) + list(x15) + list(x20))


x_plot = np.linspace(x_min , x_max)
y_plot5 = np.polyval(p5 , x_plot)
y_plot10 = np.polyval(p10 , x_plot)
y_plot15 = np.polyval(p15 , x_plot)
y_plot20 = np.polyval(p20 , x_plot)



fig, axs = plt.subplots(2, 2,figsize=(15,10))

axs[0, 0].plot(x_plot, y_plot5, 'tab:orange')
axs[0, 0].set_title('P5')
axs[0, 1].plot(x_plot, y_plot10, 'tab:red')
axs[0, 1].set_title('P10')
axs[1, 0].plot(x_plot, y_plot15, 'tab:green')
axs[1, 0].set_title('P15')
axs[1, 1].plot(x_plot, y_plot20, 'tab:blue')
axs[1, 1].set_title('P20')

for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')
fig.tight_layout(pad=2.0)

x = np.linspace(-1, 1, num= 1000, endpoint=True)
y = 1 / (1+ 25*(x**2))
p = np.polyfit(x , y , len(x) - 1)
y_plot = np.polyval(p, x_plot)
fig2, axs2 = plt.subplots(figsize=(15,5))
axs2.plot(x_plot, y_plot)
axs2.set_title('Function')
fig2.tight_layout(pad=2.0)
```
